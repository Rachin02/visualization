## Stem plot
## linefmt = line kmn type er hobe dotted or dashes or solid and kon color
## markerfmt = point kmn color and type er hobe
## basefmt = base line kon color er hobe
## bottom = frequency bottom e koto theke suru hobe
## orientation = kon angle e graph hobe vertical or horizontal
## label = color with name

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7]
y = [4,3,6,1,4,2,3]
plt.stem(x,y,linefmt = '-.r',markerfmt = '*',basefmt = 'g',bottom = 2,orientation = 'horizontal',label = 'star')
plt.legend(loc = 2)
plt.show()
```
