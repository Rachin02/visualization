### vert = True/False / boxplot vertical naki horizontal hobe 
### notch = True/False / box shap change kora hoy
### widths = box er width set kora hoy
### tick_labels = [] / boxplot er name niche show kore
### patch_artist = True/False / box will filled with color
### showmeans = True/False / mean show kore
### whis  = value / eita diye whisker line ke outlier porjonto extend kora hoy
### sym = color+icon / outlier er design change kore
### boxprops = dict() / box plot er design change kore
### capprops = dict() / min ,max line er design change kore
### whiskerprops = dict() whisker line er color change kore

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70,120]
plt.boxplot(x,vert = True, patch_artist = True, showmeans = True,notch = False, widths = .4, sym ='r*',whis = 1.6,tick_labels=['C++'],
           boxprops = {'color':'r'},capprops = {'color':'y'},whiskerprops = {'color':'g'})
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
u = [20,50,10,30,53,45,95]
y = [x,u]
plt.boxplot(y,tick_labels=['c','p'])
plt.show()
```
