#### pie plot
#### explode = slice kora bujhay jmn c++ slice kora pic chart theke
#### shadow = make a shadow of the graph
#### radius = make pie chart big
#### autopct = show percentage inside the chart
#### labeldistance = pic chart theke label er distance koto dure hobe
#### startangle = rotation pie chart. koto degree theke start hobe
#### textprops = text er design
#### counterclock = rotation of the pic chart clock wise or not
#### wedgeprops = border er design
#### rotatelabels = label er title rotate kore

```python
import matplotlib.pyplot as plt
```

```python
x = ['c','java','c++','python']
y = [10,40,30,20]
ex = [0.2,0.0,0.0,0.0]
plt.pie(y,autopct = '%1.01f%%',shadow = True, explode = ex,radius = 1.2,startangle = 290, labels = x,labeldistance = 1.2,
       rotatelabels = True,counterclock = True, textprops = {'fontsize':15},wedgeprops = {'edgecolor':'black','linewidth':1} )
plt.legend(loc=2)
plt.show()
```
