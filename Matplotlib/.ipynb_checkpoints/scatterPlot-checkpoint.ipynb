### scatte plot
### scatter() function diye scatter diagram show kore
### marker = diye style change kore point gulor
### s = size of the point
### colorbar() diye color range chart add kore

```python
import matplotlib.pyplot as plt
```

```python
day = [1,2,3,4,5,6,7]
people = [3,4,1,6,9,8,5]
c = ['r','b','g','y','g','r','g']
plt.scatter(day,people,color =c, s =200,marker = '*',edgecolor = 'black',linewidth = 1)
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
people = [3,4,1,6,9,8,5]
animal = [5,6,3,7,1,9,5]
c = ['r','b','g','y','g','r','g']
plt.scatter(day,people,marker = '*',color =c, s =200)
plt.scatter(day,animal, color='r',alpha = 0.3)
t = plt.colorbar()
t.set_label('color Bar')
plt.xlabel('hi')
plt.ylabel('hello')
plt.title('ok')
plt.show()
```
