### ylabel = y axis e titel show kore
### plt.title = full diagram er title
### edgecolor = border color
### linewidth = border width
### alpha = bar color er saturation baray komay
### label = multiple bar thakle kono specific bar er color mean kore
### leged = multiple bar thakle protita bar er jonno spearate color code show kore
### fontsize = font er size change kor

```python
import matplotlib.pyplot as plt
```

```python
lan = ['c','C++','python','java']
male = [80,92,87,75]
female =[67,72,90,85]
```

```python
plt.bar(lan,male,width = 0.6,edgecolor = 'black',linewidth = 2,color = ['r','b','y','g'],alpha = 0.7)
plt.xlabel('language')
plt.ylabel('count')
plt.title('Analysis male',fontsize = 20)
plt.show()
```

```python
import numpy as np
m = np.arange(len(lan))
w = 0.3
f = [i+w for i in m]

plt.bar(m, male, color = 'r',width = w,linewidth = 2, edgecolor = 'black',label = 'Male')
plt.bar(f, female, color = 'b',width = w, linewidth = 2, edgecolor = 'black',label = 'feMale')
plt.xticks(m+w/2,lan)
plt.legend()
plt.show()
```

```python
import numpy as np
m = np.arange(len(lan))
w = 0.3
f = [i+w for i in m]

plt.barh(m,male,color = 'g',height = w,edgecolor = 'black',label = 'Male')
plt.barh(f,female,color = 'orange',height = w,label = 'feMale')
plt.yticks(m+w/2,lan)
plt.legend()
plt.show()
```
