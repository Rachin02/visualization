```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,3,5,1]
```

```python
plt.plot(x,y)
plt.xticks(x,labels=['c','g','h','r','s'])
plt.yticks(y)
plt.xlim(0,8)
plt.ylim(0,8)
plt.text(2,2,'C++')
plt.annotate('c++',xy = (2.3,2.3),xytext = (6,6),arrowprops= {'facecolor':'r'})
plt.show()
```
