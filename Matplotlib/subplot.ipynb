```python
import matplotlib.pyplot as plt
```

```python
lan = ['c','C++','python','java']
male = [80,92,87,75]
female =[67,72,90,85]
x = [10,20,30,40,50,60,70,120]


plt.subplot(3,3,1)
plt.bar(lan,male,edgecolor='black',color = ['r','b','g','y'],linewidth = 2,width =0.6,alpha = 0.7)

plt.subplot(3,3,2)
plt.hist(male, color='y', rwidth = 0.9, edgecolor = 'black',bins = 2,label ='hist',histtype = 'barstacked',cumulative = -1)
plt.axvline(84,color = 'r',label='risk')
plt.grid()
plt.legend(loc=4)

plt.subplot(3,3,3)
e = [0.0,0.1,0.0,0.0]
plt.pie(male,explode = e, shadow = True, radius = 1.1, autopct = '%0.1f%%',labels = lan,labeldistance = 1.2,rotatelabels = True,
       startangle =240,counterclock = True, textprops = {'fontsize':10}, wedgeprops = {'edgecolor':'b','linewidth':1})

plt.subplot(3,3,4)
plt.stem(male,female, linefmt = '.-r',markerfmt = '*',basefmt = 'b',bottom = 40,orientation = 'horizontal')



plt.subplot(3,3,5)
plt.boxplot(x,notch = False,patch_artist = True,showmeans = True, whis = 1,sym = 'r*', boxprops = {'color':'r'},capprops={'color':'g'},
           whiskerprops = {'color':'y'},vert =False,widths = 0.5,tick_labels = ['c'])


plt.subplot(3,3,6)
plt.scatter(male,female, marker='*',s = 100,color = 'b')

plt.subplot(3,3,7)
plt.fill_between(x = [82,88] , y1 = 50, y2 = 60)
plt.plot(male,female)

plt.subplot(3,3,8)
plt.text(80,75,'c')
plt.annotate('got',xy = (84,78),xytext = (95,95),arrowprops = {'facecolor':'r'})
plt.plot(male,female)

plt.savefig('d1.jpg',dpi = 2000,transparent = True)
# plt.savefig('d1.jpg',dpi = 2000,facecolor = 'b',transparent = True,bbox_inches='tight')

plt.show()
```
