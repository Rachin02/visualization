```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../dengue_dataset.csv')
```

```python
df['Age'].head(3)
```

```python
l = [10,20,30,40,50,60,70,80]
plt.hist(df['Age'],edgecolor = 'black',linewidth = 2,color = 'orange',alpha = 0.7,rwidth = .7,bins = l,label ='age')
plt.xlabel('age')
plt.ylabel('count')
plt.title('age count')
plt.legend()
plt.show()
```

```python
plt.hist(df['Age'],edgecolor = 'black',linewidth = 1,rwidth = 0.8,cumulative = -1)
plt.show()
```

```python
plt.hist(df['Age'],edgecolor = 'black',linewidth = 1,rwidth = 0.8,bottom = 15,align = 'left',histtype = 'step')
plt.show()

#bar , step , stepfilled , barstacked
```

```python
plt.hist(df['Age'],edgecolor = 'black',linewidth = 1,rwidth = 0.8,bottom = 15,align = 'left',orientation = 'horizontal')
plt.show()
```

```python
plt.hist(df['Age'],edgecolor = 'black',linewidth = 1,rwidth = 0.8,bottom = 15,align = 'left',log= True,bins = 8)
plt.show()
```

```python
plt.hist(df['Age'],color = 'r',alpha =0.8, edgecolor = 'black',linewidth = 1,rwidth = 0.8)
plt.axvline(75,color = 'b',label = 'risk line')
plt.grid()
plt.legend()
plt.show()
```
