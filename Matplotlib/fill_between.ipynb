```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4,5,6]
y = [2,3,4,5,6,7]
```

```python
plt.fill_between(x,y)
plt.plot(x,y)
plt.show()
```

```python
plt.fill_between(x =[3,4],y1 = 3,y2 = 5, color = 'r')
plt.plot(x,y)
plt.show()
```

```python
plt.fill_between(x,y, where =(x1>=3)&(x2<=5) , color = 'g')
plt.plot(x,y)
plt.show()
```
