```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
a = np.linspace(1,20,20).reshape(4,5)
a
```

```python
plt.rcParams['figure.figsize']=(8,5)
```

```python
sns.heatmap(a,annot= True,vmin= 0,vmax = 30,cmap='PuOr')
```
