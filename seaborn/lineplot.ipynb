# line plot
x,y  = kon kon column niye kaj korbo

data = dataset er name

hue = used to group data by a categorical variable and assign different colors to each group.\

marker = change data point

palette = change the color for lines

style = change the line style


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,4,1,5,6,7]
df = pd.DataFrame({'id':x,'age':y})
```

```python
df
```

```python
sns.lineplot(x = df['id'],y = df['age'])
plt.show()
```

```python
a = sns.load_dataset('iris')
a.head(2)
```

```python
sns.lineplot(x = a.sepal_length, y = a.sepal_width, hue = a['species'],style = a['species'], palette = 'Accent', 
            )
```
