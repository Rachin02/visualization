# Histplot
kde = a curve line

bins = number of bar

log/log_scale = True/False

color = change color



```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../dengue_dataset.csv')
```

```python
df['Age'].head(4)
```

```python
plt.rcParams['figure.figsize'] = (8,4)
sns.set_style('darkgrid')
sns.histplot(df.Age, color = 'r',bins = 15, kde = True, log_scale = True)
```
