```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head(3)
```

```python
plt.figure(figsize = (6,4))
sns.set_style('darkgrid')
sns.barplot(x =df.day, y = df.total_bill, hue = df.sex, errorbar = ('ci',1))
```

```python
f = sns.FacetGrid(df, col = 'day', hue = 'sex')
f.map(plt.bar,'total_bill','tip',edgecolor = 'r', linewidth = 1).add_legend()
plt.show()
```
