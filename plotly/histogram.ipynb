```python
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = px.data.tips()
```

```python
df.head()
```

```python
fig = px.histogram(x = df['tip'], color = df['sex'],text_auto = True)
fig.show()
```

```python
fig = px.histogram(df,x= 'day', color ='sex')
fig.show()
```

```python
fig =px.histogram(df,x = 'total_bill',color = 'sex',nbins= 50)
fig.update_layout(bargap = .2)
fig.show()
```

```python
fig =px.histogram(df,x = 'total_bill',color = 'sex',nbins= 50,histnorm ='probability')
fig.update_layout(bargap = .2)
fig.show()
```

```python
fig =px.histogram(df,x = 'day', y = 'tip', histfunc='avg',  text_auto = True, pattern_shape = 'smoker')
fig.update_layout(bargap = .2)
fig.show()
```
