```python
import plotly.express as px
import pandas as pd
```

```python
df = px.data.gapminder()
df.head()
```

```python
df = df.query("country == 'Canada'")
```

```python
fig = px.line(df, x = 'year', y = 'lifeExp', title = 'canada life expense')
fig.show()
```

```python
fig = px.line(df, x ='year', y = 'lifeExp', text = 'year' )
fig.show()
```

```python
df = px.data.gapminder().query("continent == 'Oceania'")
df.head()
```

```python
fig = px.line(df, x = 'year', y = 'lifeExp', color = 'country', markers = True)
fig.show()
```

```python
fig = px.line(df, x = 'date', y = 'GOOG', markers = True)
fig.show()
```

```python
df = px.data.gapminder().query("country in ['Canada', 'Australia']")
```

```python
fig = px.line(df, x = 'year', y = 'gdpPercap', color = 'country', markers = True, text = 'year')
fig.show()
```
