```python
import plotly.express as px
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../dengue_dataset.csv')
```

```python
df.head(5)
```

```python
fig = px.bar(x = df['Location'], color = df['Location'])
fig.show()
```

```python
data = df.groupby(df['Location']).size().rename('Count').reset_index()
data
```

```python
d = px.bar(x =data['Location'], y = data['Count'], color = data['Location'], text = data['Count'], 
          hover_data = [data['Location']])
d.show()
```

```python
data = px.data.gapminder().query("country == 'Canada'")
data
```

```python
fig = px.bar(x = data['year'], y = data['pop'], color = data['year'], text = data['pop'],
            hover_data = [data['lifeExp'],data['iso_num']], height = 500, labels = {'hover_data_0':'lifeExp'})
fig.show()
```
