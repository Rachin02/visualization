```python
import plotly.express as px
import pandas as pd
```

```python
df = px.data.iris()
df.head()
```

```python
fig = px.scatter(df, x = 'sepal_length', y = 'sepal_width', color= 'species', symbol = 'species', size = 'petal_length')
fig.show()
```

```python
df = px.data.tips()
df.head()
```

```python
fig = px.scatter(df, x = 'total_bill', y = 'tip', color = 'smoker', facet_col = 'sex', facet_row = 'time')
fig.show()
```

```python
fig = px.scatter(df , x= 'total_bill', y = 'tip', color = 'sex', marginal_x = 'histogram', marginal_y = 'box')
fig.show()
```
