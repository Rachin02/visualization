```python
import plotly.express as px
import pandas as pd
```

```python
df= px.data.tips()
df.head()
```

```python
fig = px.box(df, y = 'total_bill')
fig.show()
```

```python
fig = px.box(df , x = 'time', y = 'total_bill' , points = 'all', color = 'sex', hover_data = ['smoker'] )
fig.show()
```

```python
fig = px.box(df , x = 'time', y = 'total_bill' , points = 'outliers', notched = True, color = 'sex' )
fig.show()
```

```python
fig = px.strip(df, x = 'smoker', y ='total_bill')
fig.show()
```
